```python
import folium
def display_map():
    center_of_theatre = (43.0719, -76.1703)
    map2 = folium.Map(location = center_of_theatre, zoom_start = 15)
    postion = (43.0719, -76.1703)
    marker = folium.Marker(location = postion, popup = 'Destiny')
    map2.add_child(marker)
    return map2
display_map()
```
